## Matplotlib


```python
import matplotlib.pyplot as plt
```

## Line Plot


```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.title("Sample Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
```

```python
plt.plot(x, y, color="red", linestyle="--", marker="o", markersize=4, linewidth=1)
```

## Multiple plots


```python
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 5, 10]

plt.figure(figsize=(10, 5))
plt.subplot(2, 2, 1)
plt.plot(x, y1, label="Line 1", linestyle="--", marker="o", markersize=4, linewidth=1)
plt.title("Line 1")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.subplot(2, 2, 2)
plt.plot(x, y2, label="Line 2", color="orange", marker="o")
plt.title("Line 2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.subplot(2, 2, 3)
plt.plot(x, y1, label="Line 1", linestyle="--", marker="o", markersize=4, linewidth=1)
plt.title("Line 1")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.subplot(2, 2, 4)
plt.plot(x, y2, label="Line 2", color="orange", marker="o")
plt.title("Line 2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.tight_layout()
```

## Bar Plot


```python
categories = ["A", "B", "C", "D", "E"]
values = [3, 7, 5, 6, 4]

plt.bar(categories, values, color="skyblue", edgecolor="black")
plt.title("Sample Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.xticks(rotation=45)
plt.yticks(range(0, 8, 1))
plt.grid(axis="y")
```

## Histogram

### represent distribution of dataset, they divide data into bins and count number of data points in each bin


```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5, 6, 7, 8]

plt.hist(data, bins=8, color="skyblue", edgecolor="black")
```

## Pie Chart


```python
labels = ["Category A", "Category B", "Category C", "Category D"]
sizes = [30, 40, 50, 20]
colors = ["gold", "lightcoral", "lightskyblue", "lightgreen"]
explode = (0.0, 0, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%")
```

## Practical - Sales Data Visualization


```python
import pandas as pd
```

```python
sales_data = pd.read_csv("sales_data.csv")
sales_data.head()
```

```python
sales_data.dtypes
```

```python
# plot total revenue by product category
total_sales_by_product = sales_data.groupby(["Product Category"])["Total Revenue"].sum()
total_sales_by_product
```

```python
total_sales_by_product.plot(
    kind="bar",
    color="skyblue",
    edgecolor="black",
    xlabel="Product Category",
    ylabel="Total Revenue",
    title="Total Revenue by Product Category",
    grid=True,
)
```

```python
# plot total revenue over time
total_sales_over_time = sales_data.groupby(["Date"])["Total Revenue"].sum()
total_sales_over_time
```

```python
total_sales_over_time.plot(
    kind="line",
    color="skyblue",
    xlabel="Product Category",
    ylabel="Total Revenue",
    title="Total Revenue by Product Category",
)
```
